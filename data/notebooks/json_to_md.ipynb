```python
import os
import json
import re
import frontmatter
```

```python
DIRNAME = os.path.abspath('')
WWW_PATH = os.path.join(DIRNAME, '../../www')
```

```python
for collection in [
    {"plural": "cities", "singular": "city"}, 
    {"plural": "ships", "singular": "ship"},
    {"plural": "routes", "singular": "route"},
    {"plural": "companies", "singular": "company"},
]:
    for filename in os.listdir(os.path.join(WWW_PATH, f"_data_tmp/{collection['plural']}")):
        json_path = os.path.join(WWW_PATH, f"_data_tmp/{collection['plural']}/{filename}")
        with open(json_path, "r") as f:
            data = json.load(f)
        md_path = os.path.join(WWW_PATH, f"{collection['plural']}/{data['slug']}.md")
        with open(md_path, "wb") as f:
            frontmatter.dump(frontmatter.Post("", layout=collection["singular"], tags=[collection["singular"]], **data), f)
        print(f"converted {json_path} to {md_path}")
```
